```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

def decision_surface_draw(arr, tgt):
    # Parameters
    n_classes = 3
    plot_colors = "ryb"
    plot_step = 0.02

    numpy_arr = np.array(arr)
    numpy_tgt = np.array(tgt)

    feature_names = ['postura_profissional', 'atuacao_profissional', 'autoavaliacao_aluno', 'periodo']
    tgt_names = ['<7', '>= 7 && <9', '>=9']

    fig_list = []  # Lista para armazenar as figuras

    for pairidx, pair in enumerate([[0, 1], [0, 2], [0, 3],
                                    [1, 2], [1, 3], [2, 3]]):

        # We only take the two corresponding features
        X = numpy_arr[:, pair]
        y = numpy_tgt

        # Train
        clf = DecisionTreeClassifier().fit(X, y)

        # Plot the decision boundary
        plt.figure()
        fig_list.append(plt.gcf())  # Adiciona a figura atual à lista

        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                             np.arange(y_min, y_max, plot_step))
        plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        cs = plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu)

        plt.xlabel(feature_names[pair[0]])
        plt.ylabel(feature_names[pair[1]])

        # Plot the training points
        for i, color in zip(range(n_classes), plot_colors):
            idx = np.where(y == i)
            plt.scatter(X[idx, 0], X[idx, 1], c=color, label=tgt_names[i],
                        cmap=plt.cm.RdYlBu, edgecolor='black', s=15)

        plt.suptitle("Decision surface of a decision tree using paired features")
        plt.legend(loc='lower right', borderpad=0, handletextpad=0)
        plt.axis("tight")

    return fig_list  # Retorna a lista de figuras

# Exemplo de uso com dados do conjunto de dados Iris
iris = load_iris()
data = iris.data
target = iris.target
figures = decision_surface_draw(data, target)

plt.show()  # Mostra todas as figuras plotadas
```
